# Theoretical Questions:

What is hypothesis testing in statistics?

Hypothesis testing in statistics is a method used to make inferences or draw conclusions about a population based on sample data. It involves formulating two competing hypotheses: the null hypothesis (H0), which represents a statement of no effect or no difference, and the alternative hypothesis (H1 or Ha), which represents a statement of an effect or a difference. The process includes:
1. **Formulating Hypotheses**: Define the null and alternative hypotheses.
2. **Choosing a Significance Level**: Select a significance level (alpha, typically 0.05) that determines the threshold for rejecting the null hypothesis.
3. **Collecting Data**: Gather sample data relevant to the hypotheses.
4. **Calculating a Test Statistic**: Compute a test statistic based on the sample data, which measures how far the sample statistic is from the null hypothesis.
5. **Finding the P-value**: Calculate the p-value, which indicates the probability of observing the test statistic or something more extreme if the null hypothesis is true.
6. **Making a Decision**: Compare the p-value to the significance level. If the p-value is less than or equal to the significance level, reject the null hypothesis; otherwise, do not reject it.

What is the null hypothesis, and how does it differ from the alternative hypothesis?



The null hypothesis (H0) is a statement that there is no effect, no difference, or no relationship in the population being studied. It serves as a default position that indicates no change or no association. The alternative hypothesis (H1 or Ha), on the other hand, is a statement that contradicts the null hypothesis, suggesting that there is an effect, a difference, or a relationship in the population. The key difference is that the null hypothesis assumes no effect or relationship, while the alternative hypothesis posits that such an effect or relationship exists.


What is the significance level in hypothesis testing, and why is it important?

The significance level (alpha, typically set at 0.05) in hypothesis testing is the threshold for determining whether to reject the null hypothesis. It represents the probability of making a Type I error, which occurs when the null hypothesis is incorrectly rejected when it is actually true. The significance level is important because it helps control the risk of false positives in hypothesis testing, allowing researchers to make informed decisions about the evidence against the null hypothesis. A lower significance level reduces the likelihood of Type I errors but increases the risk of Type II errors (failing to reject a false null hypothesis). Thus, choosing an appropriate significance level balances these risks based on the context of the study.


What does a P-value represent in hypothesis testing?

A P-value in hypothesis testing represents the probability of observing the test statistic or something more extreme, given that the null hypothesis is true. It quantifies the strength of evidence against the null hypothesis. A low P-value (typically less than or equal to the significance level, such as 0.05) suggests that the observed data is unlikely under the null hypothesis, leading to its rejection. Conversely, a high P-value indicates that the observed data is consistent with the null hypothesis, and there is insufficient evidence to reject it. Thus, the P-value helps researchers assess whether the sample data provides strong enough evidence to support the alternative hypothesis.

How do you interpret the P-value in hypothesis testing?

The P-value in hypothesis testing is interpreted as follows:
The P-value in hypothesis testing is interpreted as the probability of obtaining a test statistic at least as extreme as the one observed, assuming that the null hypothesis is true. It provides a measure of the strength of evidence against the null hypothesis. The interpretation of the P-value can be summarized in two main scenarios based on its comparison to the significance level (α):
1. **Low P-value (≤ α)**: If the P-value is less than or equal to the significance level (commonly set at 0.05), it indicates strong evidence against the null hypothesis. In this case, researchers reject the null hypothesis in favor of the alternative hypothesis, suggesting that there is a statistically significant effect or difference.
2. **High P-value (> α)**: If the P-value is greater than the significance level, it suggests that the observed data is consistent with the null hypothesis. In this scenario, researchers do not reject the null hypothesis, indicating insufficient evidence to support the alternative hypothesis.

What are Type 1 and Type 2 errors in hypothesis testing?

Type 1 and Type 2 errors are two types of errors that can occur in hypothesis testing:

- **Type 1 Error (False Positive)**: This error occurs when the null hypothesis is incorrectly rejected when it is actually true. In other words, it is the error of concluding that there is an effect or difference when there is none. The probability of making a Type 1 error is denoted by the significance level (alpha, α), typically set at 0.05.

- **Type 2 Error (False Negative)**: This error occurs when the null hypothesis is not rejected when it is actually false. In other words, it is the error of failing to detect an effect or difference that truly exists. The probability of making a Type 2 error is denoted by beta (β). The power of a test, which is the probability of correctly rejecting a false null hypothesis, is equal to 1 - β.

What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

A one-tailed test and a two-tailed test are two types of hypothesis tests that differ in how they assess the direction of the effect or difference being tested:

- **One-tailed Test**: A one-tailed test is used when the alternative hypothesis specifies a direction of the effect or difference. For example, if the alternative hypothesis states that a new drug is more effective than an existing one, the test will only look for evidence in that specific direction (greater than). This type of test has more statistical power to detect an effect in the specified direction but does not consider the possibility of an effect in the opposite direction.

- **Two-tailed Test**: A two-tailed test is used when the alternative hypothesis does not specify a direction, meaning it tests for the possibility of an effect in both directions (greater than or less than). For example, if the alternative hypothesis states that a new drug has a different effect compared to an existing one, the test will look for evidence of both an increase and a decrease in effectiveness. This type of test is more conservative and requires a larger effect size to achieve statistical significance but is appropriate when the direction of the effect is not known beforehand.

What is the Z-test, and when is it used in hypothesis testing?

A Z-test is a statistical test used in hypothesis testing to determine whether there is a significant difference between the means of two groups or whether a sample mean significantly differs from a known population mean. It is applicable when the following conditions are met:

1. The sample size is large (typically n > 30), or the population standard deviation is known.

2. The data is approximately normally distributed, or the Central Limit Theorem applies (which states that the sampling distribution of the sample mean approaches normality as the sample size increases).

###  Z-Score: Definition, Formula, and Role in Hypothesis Testing

---

####  What is a Z-Score?

A **Z-score** is a statistical measure that tells us **how many standard deviations** a data point is from the **mean** of the population.

---

###  Z-Score Formula

**Z = (X - μ) / σ**

Where:
- `Z` = Z-score
- `X` = Observed value (sample mean or individual value)
- `μ` = Population mean
- `σ` = Population standard deviation

---

####  Example

Suppose:
- Population mean (μ) = 100  
- Population standard deviation (σ) = 15  
- Observed value (X) = 115

**Z = (115 - 100) / 15 = 1**

> This means the value is **1 standard deviation above the mean**.

---

###  Z-Score in Hypothesis Testing

In hypothesis testing, the Z-score helps us determine how far our sample statistic is from the assumed population mean, under the null hypothesis (H₀).

---

###  Steps in Z-Test (One Sample)

1. **State Hypotheses**
   - H₀: μ = μ₀ (null hypothesis)
   - H₁: μ ≠ μ₀ (two-tailed) or μ > μ₀ / μ < μ₀ (one-tailed)

2. **Calculate the Z-score**

   **Z = (X̄ - μ) / (σ / √n)**

   Where:
   - `X̄` = sample mean
   - `μ` = population mean under H₀
   - `σ` = population standard deviation
   - `n` = sample size

3. **Find the critical Z value**
   - For α = 0.05 (two-tailed test), critical Z = ±1.96

4. **Make a decision**
   - If calculated Z > critical Z → **Reject H₀**
   - If calculated Z ≤ critical Z → **Fail to reject H₀**

---

###  What the Z-Score Represents

- A **large Z-score** (positive or negative) → strong evidence **against H₀**
- A **Z-score close to 0** → sample mean is close to population mean; **support for H₀**

---

###  Real-Life Use Cases

- A/B testing in marketing
- Clinical trials in medicine
- Manufacturing quality checks
- Educational research performance analysis


### T-Distribution: Definition, Formula, and Usage

---

#### What is the T-Distribution?

The **T-distribution** (also called Student's t-distribution) is a probability distribution that is symmetric and bell-shaped, like the normal distribution, but has **heavier tails**. This means it is more prone to producing values that fall far from its mean.

It is especially useful when the **sample size is small** and the **population standard deviation is unknown**.

---

### Formula of T-Statistic

**T = (X̄ - μ) / (s / √n)**

Where:
- `T` = t-score
- `X̄` = sample mean
- `μ` = population mean under the null hypothesis
- `s` = sample standard deviation
- `n` = sample size

---

#### When to Use the T-Distribution Instead of the Normal Distribution

Use the **T-distribution** in the following cases:

1. **Small Sample Size (n < 30)**: When working with small samples, the estimate of the standard deviation becomes less reliable. The t-distribution accounts for this uncertainty.
2. **Unknown Population Standard Deviation**: If the population standard deviation (σ) is not known and is instead estimated from the sample data.
3. **Assuming the Population is Normally Distributed**: The t-distribution assumes that the underlying population is normally distributed. For large samples, this becomes less important due to the Central Limit Theorem.

---

### Comparison with Normal Distribution

| Feature                  | T-Distribution                  | Normal Distribution               |
|--------------------------|----------------------------------|------------------------------------|
| Shape                   | Bell-shaped, heavier tails        | Bell-shaped                        |
| Use Case                | Small samples, σ unknown          | Large samples, σ known             |
| Variability             | More spread out                  | Less spread out                    |
| Degrees of Freedom      | Required                          | Not required                       |
| Approaches Normal Curve | As sample size increases          | Already fixed                      |

---

### Real-Life Applications

- Estimating the mean of a population from a small sample
- Conducting a t-test for comparing sample mean to a known value
- Paired and independent sample comparisons in research

---

### Conclusion

The T-distribution is a critical tool in inferential statistics, especially when dealing with small samples and unknown population parameters. It provides more conservative and realistic results under these conditions, making it preferable over the normal distribution in such cases.


### What is the difference between a Z-test and a T-test?

---

### Definition

Both the **Z-test** and **T-test** are statistical hypothesis tests used to determine whether there is a significant difference between sample statistics and population parameters. However, they differ based on sample size and whether the population standard deviation is known.

---

### Key Differences Between Z-Test and T-Test

| Criteria                         | Z-Test                                     | T-Test                                      |
|----------------------------------|--------------------------------------------|---------------------------------------------|
| **Population Standard Deviation**| Known                                       | Unknown                                     |
| **Sample Size (n)**              | Large (typically n ≥ 30)                   | Small (typically n < 30)                    |
| **Distribution Used**            | Normal distribution                        | T-distribution                              |
| **Test Statistic Formula**       | Z = (X̄ - μ) / (σ / √n)                     | T = (X̄ - μ) / (s / √n)                      |
| **Tail Thickness**               | Thinner tails                              | Thicker tails (to account for more variability) |
| **Use Case**                     | When population parameters are known       | When population standard deviation is unknown and estimated from sample |
| **Accuracy with Small Samples**  | Less accurate                              | More accurate                               |

---

### Explanation

- A **Z-test** assumes you know the **population standard deviation (σ)** and that the sample size is large enough for the sampling distribution to be approximately normal.
- A **T-test** is used when the **population standard deviation is unknown**, and the sample size is small. It uses the sample standard deviation to estimate variability and relies on the **t-distribution**, which adjusts for increased uncertainty.

---

### Example Use Cases

- **Z-test**: Comparing the average height of a large group of people to a known population mean, where σ is known.
- **T-test**: Testing the effectiveness of a new drug with a sample of 15 patients where σ is not known.

---

### Conclusion

Choose a **Z-test** when the population standard deviation is known and the sample size is large. Use a **T-test** when the sample size is small and the population standard deviation is unknown. As the sample size increases, the t-distribution approaches the normal distribution, making the difference between the two tests less significant.


### What is the T-test, and how is it used in hypothesis testing?

---

### Definition

A **T-test** is a statistical test used to determine whether there is a significant difference between the means of two groups or between a sample mean and a known population mean. It is especially useful when the **sample size is small (n < 30)** and the **population standard deviation is unknown**.

The T-test relies on the **t-distribution**, which adjusts for increased variability due to estimating the population standard deviation from the sample.

---

### Types of T-Tests

1. **One-Sample T-Test**: Compares the sample mean to a known population mean.
2. **Independent Two-Sample T-Test**: Compares the means of two independent groups.
3. **Paired Sample T-Test**: Compares means from the same group at two different times (e.g., before and after treatment).

---

### T-Test Formula (One-Sample)

**T = (X̄ - μ) / (s / √n)**

Where:
- `T` = t-score
- `X̄` = sample mean
- `μ` = population mean under the null hypothesis
- `s` = sample standard deviation
- `n` = sample size

---

### Steps to Perform a T-Test in Hypothesis Testing

1. **State the Hypotheses**
   - Null Hypothesis (H₀): There is no difference (e.g., μ = μ₀)
   - Alternative Hypothesis (H₁): There is a difference (e.g., μ ≠ μ₀ for a two-tailed test)

2. **Choose Significance Level (α)**
   - Common choices are 0.05 or 0.01

3. **Calculate the T-Statistic**
   - Use the formula based on the test type

4. **Find the Degrees of Freedom**
   - For one-sample t-test: df = n - 1

5. **Determine the Critical T-Value**
   - From the t-distribution table using α and df

6. **Compare T-Statistic with Critical T-Value**
   - If the absolute value of the t-statistic is greater than the critical t-value → reject H₀

7. **Draw a Conclusion**
   - Reject or fail to reject the null hypothesis based on the comparison

---

### Example Use Case

Suppose a nutritionist wants to test whether a new diet affects weight. A sample of 10 individuals loses an average of 3.2 kg with a sample standard deviation of 1.5 kg. The null hypothesis is that the diet causes no weight loss (μ = 0). A one-sample t-test would determine if the average loss is statistically significant.

---

### Conclusion

The T-test is a core tool in inferential statistics used to test hypotheses about population means when sample sizes are small and population variability is not known. It provides a flexible and robust approach to hypothesis testing in practical research scenarios.


### What is the relationship between Z-test and T-test in hypothesis testing?

---

### Overview

Both the **Z-test** and **T-test** are statistical methods used in **hypothesis testing** to assess whether there is a significant difference between sample data and a population parameter (or between two groups). They follow similar logic but differ based on **sample size** and **knowledge of the population standard deviation**.

---

### Similarities

1. **Purpose**: Both tests are used to test hypotheses about population means.
2. **Test Structure**: Both compare a test statistic (Z or T) to a critical value to decide whether to reject the null hypothesis.
3. **Assumptions**:
   - Data are randomly sampled.
   - Observations are independent.
   - Population is normally distributed or sample size is large (Central Limit Theorem applies).

---

### Differences

| Feature                        | Z-Test                                | T-Test                                  |
|-------------------------------|----------------------------------------|------------------------------------------|
| Population Standard Deviation | Known                                  | Unknown (estimated from sample)          |
| Sample Size                   | Large (n ≥ 30)                         | Small (n < 30)                            |
| Distribution Used             | Normal distribution                    | T-distribution                           |
| Accuracy for Small Samples    | Less reliable                          | More reliable due to heavier tails       |
| Degrees of Freedom            | Not applicable                         | Required (df = n - 1 or similar)         |

---

### Relationship

- **The T-distribution approaches the normal distribution** as the sample size increases.
- This means that the **T-test becomes almost identical to the Z-test** for large samples.
- In practice:
  - For **large samples**, both tests will yield very similar results.
  - For **small samples**, the T-test is preferred because it adjusts for greater uncertainty.

---

### Visual Relationship

- T-distribution has **heavier tails**, accounting for more variability.
- As degrees of freedom increase (larger sample size), the **t-distribution becomes narrower** and converges toward the **normal (z) distribution**.

---

### Conclusion

The **T-test is an extension of the Z-test** that is more suitable when the population standard deviation is unknown and/or the sample size is small. Understanding the relationship helps in choosing the correct test for accurate hypothesis testing.


### What is a confidence interval, and how is it used to interpret statistical results?

---

### Definition

A **confidence interval (CI)** is a **range of values**, derived from sample statistics, that is likely to contain the **true population parameter** (such as the mean or proportion) with a certain level of confidence.

It reflects the **uncertainty or variability** in the estimate due to sampling.

---

### Structure of a Confidence Interval

**Confidence Interval = Point Estimate ± Margin of Error**

For a population mean:

**CI = X̄ ± Z * (σ / √n)**  *(if σ is known, large sample)*  
**CI = X̄ ± t * (s / √n)**  *(if σ is unknown, small sample)*

Where:
- `X̄` = sample mean  
- `σ` = population standard deviation  
- `s` = sample standard deviation  
- `n` = sample size  
- `Z` or `t` = critical value from the Z or T distribution based on the confidence level

---

### Common Confidence Levels

| Confidence Level | Z-Value (Approx.) |
|------------------|-------------------|
| 90%              | 1.645             |
| 95%              | 1.96              |
| 99%              | 2.576             |

---

### Interpretation

A **95% confidence interval** means that if we were to take 100 random samples and compute a CI for each, about **95 of them would contain the true population parameter**.

**Important Note:**  
A confidence interval **does not mean** there's a 95% probability that the population parameter lies in this particular interval. Instead, it means we are 95% confident in the method that produced this interval.

---

### Example

Suppose a sample of 50 students has a mean test score of 72 with a sample standard deviation of 8. We want a 95% confidence interval for the true mean score.

Using the t-distribution (because σ is unknown), with df = 49:

**CI = 72 ± t * (8 / √50)**  
Assume t ≈ 2.009 (from t-table at 95% confidence with df = 49)  

**CI ≈ 72 ± 2.009 * 1.131 = (69.73, 74.27)**

So, we are 95% confident that the **true average score** lies between **69.73 and 74.27**.

---

### How It Is Used

- **Estimating unknown population parameters**
- **Quantifying uncertainty** around a sample estimate
- **Comparing groups** (e.g., overlapping intervals may indicate no significant difference)
- **Reporting results** in surveys, experiments, and trials

---

### Conclusion

Confidence intervals are a key tool in inferential statistics. They allow researchers to **draw conclusions about a population** based on sample data, while accounting for sampling error.


### What is the margin of error, and how does it affect the confidence interval?

---

### Definition

The **margin of error (MoE)** is a measure of the **maximum expected difference** between the **true population parameter** and the **sample estimate**. It reflects the **amount of uncertainty** in the estimate due to sampling variability.

It is an essential part of the confidence interval formula and helps define how **wide or narrow** the interval is.

---

### Formula

For estimating a population mean:

- **When population standard deviation (σ) is known**:

  **MoE = Z * (σ / √n)**

- **When σ is unknown (using sample standard deviation)**:

  **MoE = t * (s / √n)**

Where:
- `Z` = critical value from the Z-distribution (e.g., 1.96 for 95% confidence)
- `t` = critical value from the T-distribution (used when σ is unknown)
- `s` = sample standard deviation
- `n` = sample size

---

### Role in the Confidence Interval

The **confidence interval (CI)** is calculated as:

**CI = Point Estimate ± Margin of Error**

Therefore, the **margin of error determines the width** of the interval. A **larger MoE** results in a **wider interval**, indicating more uncertainty. A **smaller MoE** gives a **narrower interval**, indicating more precision.

---

### Factors Affecting Margin of Error

1. **Sample Size (n)**:
   - Larger sample size → smaller MoE → more precise estimate

2. **Confidence Level**:
   - Higher confidence level (e.g., 99% vs. 95%) → larger Z/t value → larger MoE → wider interval

3. **Standard Deviation (σ or s)**:
   - More variability in data → larger MoE

---

### Example

Suppose we have:
- Sample mean (X̄) = 100
- Sample standard deviation (s) = 10
- Sample size (n) = 25
- Confidence level = 95% → t ≈ 2.064 (df = 24)

**MoE = 2.064 * (10 / √25) = 2.064 * 2 = 4.128**

Then the confidence interval is:
**CI = 100 ± 4.128 = (95.872, 104.128)**

---

### Conclusion

The **margin of error quantifies uncertainty** in statistical estimates. It directly affects the **width of the confidence interval**, making it a key component in interpreting how reliable or precise a sample estimate is. Smaller margins of error lead to more confidence in the estimate's closeness to the true population value.


### How is Bayes' Theorem used in statistics, and what is its significance?

---

### Definition

**Bayes' Theorem** is a fundamental rule in probability and statistics that describes how to **update the probability of a hypothesis** when given new evidence. It forms the basis for **Bayesian inference**, a powerful statistical method.

---

### Bayes' Theorem Formula

**P(A|B) = [P(B|A) × P(A)] / P(B)**

Where:
- `P(A|B)` = Posterior probability (probability of event A given that B has occurred)
- `P(B|A)` = Likelihood (probability of event B given A is true)
- `P(A)` = Prior probability (initial belief about A)
- `P(B)` = Marginal probability (overall probability of B)

---

### Explanation of Terms

- **Prior (P(A))**: Initial belief before seeing the data.
- **Likelihood (P(B|A))**: How likely the data is under the assumption that A is true.
- **Posterior (P(A|B))**: Updated belief after considering the data.
- **Marginal (P(B))**: The total probability of observing B.

---

### Example

Suppose 1% of people have a disease. A test correctly identifies the disease 99% of the time (true positive), and incorrectly identifies it in 5% of healthy people (false positive).

Let:
- `A` = person has the disease  
- `B` = person tests positive

We want to find: **P(A|B)**

Given:
- P(A) = 0.01  
- P(B|A) = 0.99  
- P(B|¬A) = 0.05 → P(¬A) = 0.99  

Using Bayes' Theorem:

**P(A|B) = [0.99 × 0.01] / [(0.99 × 0.01) + (0.05 × 0.99)]  
= 0.0099 / (0.0099 + 0.0495)  
= 0.0099 / 0.0594 ≈ 0.1667**

So, even with a positive result, the probability that the person actually has the disease is only **~16.67%**.

---

### Applications of Bayes' Theorem

1. **Medical Diagnosis**: Estimating disease likelihood after a test result.
2. **Spam Detection**: Calculating the probability that an email is spam based on keywords.
3. **Machine Learning**: Naive Bayes classifier.
4. **Decision Making**: Updating predictions as new data becomes available.
5. **Forensic Science**: Calculating probabilities of guilt or innocence based on evidence.

---

### Significance in Statistics

- **Shifts Perspective**: From fixed parameters to **probabilities of hypotheses**.
- **Foundation of Bayesian Statistics**: Unlike frequentist statistics, it incorporates prior knowledge.
- **Adaptive Learning**: Useful in real-time systems and iterative learning environments.
- **More Intuitive**: Especially in problems with incomplete or evolving information.

---

### Conclusion

**Bayes' Theorem** provides a structured way to **revise beliefs** in light of new evidence. Its use in modern statistics, machine learning, and real-world decision-making showcases its deep significance and versatility in handling uncertainty.


### What is the Chi-square distribution, and when is it used?

---

### Definition

The **Chi-square (χ²) distribution** is a probability distribution used in statistics to describe the distribution of the **sum of the squares of independent standard normal variables**. It is **skewed to the right** and is used primarily in **categorical data analysis**.

It is defined only for **positive values** and is determined by its **degrees of freedom (df)**.

---

### Formula

If Z is a standard normal variable, then:

**χ² = Z₁² + Z₂² + ... + Zₖ²**  
Where `k` is the degrees of freedom.

---

### Key Characteristics

- **Asymmetrical (right-skewed)**  
- **Only positive values** (χ² ≥ 0)  
- **Shape depends on degrees of freedom**  
  - For small df: highly skewed  
  - For large df: approaches normal distribution

---

### Common Uses of Chi-square Distribution

1. **Chi-square Test of Independence**  
   - Tests whether two categorical variables are independent  
   - Example: Is gender associated with product preference?

2. **Chi-square Goodness-of-Fit Test**  
   - Tests whether a sample matches an expected distribution  
   - Example: Does a die show equal frequencies for all faces?

3. **Chi-square Test for Homogeneity**  
   - Compares the distribution of a categorical variable across different groups

4. **Confidence Intervals for Variance**
   - Used in constructing confidence intervals for population variance or standard deviation

---

### Example – Chi-square Test of Independence

Suppose a researcher surveys 100 people to study the relationship between **gender** and **ice cream preference**. The observed frequencies are compared to expected frequencies (assuming no association), and a **χ² statistic** is calculated.

If the calculated χ² value is greater than the critical value from the chi-square distribution table (based on degrees of freedom and significance level), we **reject the null hypothesis** and conclude there is a significant association between gender and ice cream preference.

---

### Formula for Chi-square Test Statistic

**χ² = Σ [(Oᵢ - Eᵢ)² / Eᵢ]**

Where:
- `Oᵢ` = observed frequency
- `Eᵢ` = expected frequency

---

### Conclusion

The **Chi-square distribution** is a powerful tool in **categorical data analysis**, especially for testing relationships and distributions. It allows researchers to make inferences about populations based on sample data in a non-parametric way.


### What is the Chi-square goodness of fit test, and how is it applied?

---

### Definition

The **Chi-square goodness of fit test** is a **statistical test** used to determine whether a **sample data set matches an expected distribution**. It compares the observed frequencies in categories to the frequencies we would expect if a certain hypothesis about the distribution were true.

---

### Purpose

To test whether the observed distribution of a **single categorical variable** fits a known or expected distribution (like uniform, binomial, or any custom distribution).

---

### Hypotheses

- **Null Hypothesis (H₀)**: The observed data follows the expected distribution.
- **Alternative Hypothesis (H₁)**: The observed data does not follow the expected distribution.

---

### Formula

**χ² = Σ [(Oᵢ - Eᵢ)² / Eᵢ]**

Where:
- `Oᵢ` = observed frequency in category i  
- `Eᵢ` = expected frequency in category i  
- The sum is taken over all categories.

---

### Conditions to Use

- Data must be **counts or frequencies**.
- Categories must be **mutually exclusive**.
- Expected frequency in each category should be **at least 5**.

---

### Degrees of Freedom

**df = (number of categories - 1)**

---

### Example

Suppose a six-sided die is rolled 60 times, and the outcomes are:

| Face | Observed (Oᵢ) | Expected (Eᵢ) |
|------|---------------|---------------|
| 1    | 8             | 10            |
| 2    | 9             | 10            |
| 3    | 12            | 10            |
| 4    | 11            | 10            |
| 5    | 10            | 10            |
| 6    | 10            | 10            |

**Step 1: Calculate χ²**  
χ² = Σ [(Oᵢ - Eᵢ)² / Eᵢ]  
= (8−10)²/10 + (9−10)²/10 + (12−10)²/10 + (11−10)²/10 + (10−10)²/10 + (10−10)²/10  
= 4/10 + 1/10 + 4/10 + 1/10 + 0 + 0 = 1.0

**Step 2: Find Critical Value**  
- df = 6 - 1 = 5  
- For α = 0.05, critical χ² = 11.07 (from table)

**Step 3: Decision**  
Since 1.0 < 11.07, we **fail to reject H₀**.  
Conclusion: The die appears to be fair.

---

### Applications

- Checking if a die is fair
- Testing expected market share vs. actual sales
- Verifying distribution of survey responses

---

### Conclusion

The **Chi-square goodness of fit test** is a key tool in determining whether observed categorical data aligns with a theoretical distribution. It provides a quantitative method for validating assumptions in categorical data modeling.


### What is the F-distribution, and when is it used in hypothesis testing?

---

### Definition

The **F-distribution** is a **continuous probability distribution** that arises frequently in **hypothesis testing**, particularly when comparing **two variances** or assessing the **overall significance in regression models**.

It is defined as the ratio of two scaled chi-square distributions and is **skewed to the right**.

---

### Formula

If:
- U₁ ~ χ²(df₁)
- U₂ ~ χ²(df₂)

Then the F-distribution is defined as:

**F = (U₁ / df₁) / (U₂ / df₂)**

Where:
- `df₁` and `df₂` are degrees of freedom for the numerator and denominator, respectively.

---

### Properties

- Only defined for **positive values** (F ≥ 0)
- **Right-skewed**, especially for smaller degrees of freedom
- As degrees of freedom increase, the distribution becomes more symmetric

---

### When Is It Used?

#### 1. **Comparing Two Variances**
   - **F-test for Equality of Variances**
   - Tests whether two populations have equal variances

#### 2. **ANOVA (Analysis of Variance)**
   - Most common application
   - Compares means of three or more groups by analyzing variances
   - F-statistic helps determine if at least one group mean differs significantly

#### 3. **Regression Analysis**
   - Tests the overall **significance of the regression model**
   - Determines whether the explanatory variables collectively have a statistically significant effect on the response variable

---

### Example – ANOVA Scenario

Suppose we are testing whether the mean productivity differs between 3 different training programs.

- H₀: μ₁ = μ₂ = μ₃ (all means are equal)
- H₁: At least one mean is different

We compute an **F-statistic** using ANOVA, and compare it to a critical value from the **F-distribution table**.

- If **F_calculated > F_critical**, we reject H₀ and conclude that at least one training method has a different mean productivity.

---

### Conclusion

The **F-distribution** plays a central role in hypothesis tests involving **variances** and **multiple group comparisons**, such as:
- **F-tests**
- **ANOVA**
- **Multiple regression**

It helps assess whether observed differences are statistically significant by comparing **ratios of variances**.


### What is an ANOVA test, and what are its assumptions?

---

### Definition

**ANOVA (Analysis of Variance)** is a **statistical test** used to determine whether there are any **statistically significant differences between the means** of **three or more independent groups**.

It analyzes the **variability between group means** and compares it to the **variability within groups** to assess whether any observed differences are likely due to random chance or actual group effects.

---

### Types of ANOVA

1. **One-Way ANOVA**  
   - Used to test differences in means across one factor (independent variable).  
   - Example: Comparing exam scores across 3 teaching methods.

2. **Two-Way ANOVA**  
   - Tests the effect of **two independent variables** and their interaction.  
   - Example: Comparing scores based on both teaching method and gender.

---

### Hypotheses in One-Way ANOVA

- **Null Hypothesis (H₀)**: μ₁ = μ₂ = μ₃ = ... = μₖ (all group means are equal)
- **Alternative Hypothesis (H₁)**: At least one group mean is different

---

### Test Statistic

**F = Variance between groups / Variance within groups**

A high F-value suggests that the group means are not all equal.

---

### Assumptions of ANOVA

1. **Independence**  
   - Observations are independent both **within** and **between** groups.

2. **Normality**  
   - Data in each group should be approximately **normally distributed**.

3. **Homogeneity of Variance (Homoscedasticity)**  
   - All groups should have **similar variances**.

---

### Example

Suppose a researcher wants to compare the effectiveness of 3 diets (A, B, and C) on weight loss. After collecting data, they apply a **one-way ANOVA** to test whether the average weight loss differs significantly among the three diets.

If the **p-value < 0.05**, the researcher rejects the null hypothesis and concludes that at least one diet leads to a significantly different weight loss.

---

### When to Use ANOVA

- When comparing **3 or more group means**
- When the assumptions of independence, normality, and equal variances are met
- In experimental design, quality control, and behavioral sciences

---

### Conclusion

**ANOVA** helps determine if differences between group means are **statistically significant** by comparing variances. It is a robust method widely used in research, provided its **key assumptions** are checked and met before applying the test.


### What are the different types of ANOVA tests?

---

### Definition Recap

**ANOVA (Analysis of Variance)** is used to compare **means** across groups by analyzing **variance**. Depending on the number of independent variables (factors) and how the groups are structured, there are different types of ANOVA tests.

---

### 1. **One-Way ANOVA**

- **Purpose**: Compares the means of **three or more independent groups** based on **one independent variable (factor)**.
- **Example**: Comparing average test scores of students taught by 3 different teachers.

**Hypotheses**:
- H₀: μ₁ = μ₂ = μ₃ = ... = μₖ  
- H₁: At least one mean is different

---

### 2. **Two-Way ANOVA**

- **Purpose**: Examines the influence of **two independent variables** (factors) on a **dependent variable**, and also tests for **interaction effects** between the two factors.
- **Example**: Studying how **teaching method** and **gender** affect student performance.

**Types of effects tested**:
- Main effect of Factor A
- Main effect of Factor B
- Interaction effect between A and B

---

### 3. **Repeated Measures ANOVA**

- **Purpose**: Used when the **same subjects** are measured **multiple times** under different conditions or time points.
- **Example**: Measuring blood pressure of patients **before**, **during**, and **after** treatment.

**Advantage**:
- Controls for individual variability
- Requires fewer subjects

---

### 4. **Multivariate ANOVA (MANOVA)**

- **Purpose**: Extension of ANOVA when there are **two or more dependent variables**.
- **Example**: Testing the effect of a training program on both **test scores** and **reaction times** simultaneously.

**Benefits**:
- Helps identify effects across multiple outcomes
- Controls for correlation between dependent variables

---

### 5. **ANCOVA (Analysis of Covariance)**

- **Purpose**: Combines ANOVA and regression by adjusting for **covariates** (continuous variables that may affect the outcome).
- **Example**: Comparing exam scores across schools while controlling for **prior GPA**.

**Key Feature**:
- Removes the impact of covariates to better isolate the effect of the main factor(s)

---

### Summary Table

| Type                     | Factors       | Repeated Measures | Multiple DVs | Adjusts for Covariates |
|--------------------------|---------------|-------------------|--------------|-------------------------|
| One-Way ANOVA            | 1             | No                | No           | No                      |
| Two-Way ANOVA            | 2             | No                | No           | No                      |
| Repeated Measures ANOVA  | 1+            | Yes               | No           | No                      |
| MANOVA                   | 1+            | No (or Yes*)      | Yes          | No                      |
| ANCOVA                   | 1+            | No                | No           | Yes                     |

---

### Conclusion

Choosing the right **type of ANOVA** depends on:
- The **number of independent variables**
- Whether measurements are **repeated**
- The **number of dependent variables**
- The presence of **covariates**

Each type allows researchers to tailor the analysis to the structure and goals of their experiment.


### What is the F-test, and how does it relate to hypothesis testing?

---

### Definition

The **F-test** is a statistical test that compares the **variances** of two or more groups to determine whether they are **significantly different**. It uses the **F-distribution** to evaluate the **ratio of variances** and is most commonly used in **ANOVA** and in testing the **equality of variances**.

---

### Purpose

The F-test is used in hypothesis testing to:
- Compare **two population variances**
- Assess the **overall significance** in **regression analysis**
- Determine whether the **group means** differ significantly in **ANOVA**

---

### Formula

**F = Variance₁ / Variance₂**

Where:
- Variance₁ is the variance of the first sample (typically the larger)
- Variance₂ is the variance of the second sample

The resulting F-value is compared against a critical value from the F-distribution table based on degrees of freedom.

---

### Hypotheses

For variance comparison:
- **Null Hypothesis (H₀)**: σ₁² = σ₂² (the two variances are equal)
- **Alternative Hypothesis (H₁)**: σ₁² ≠ σ₂² (the variances are different)

For ANOVA (testing group means):
- **Null Hypothesis (H₀)**: All group means are equal
- **Alternative Hypothesis (H₁)**: At least one group mean is different

---

### Assumptions

- The populations are normally distributed.
- The samples are independent.
- The data is continuous.

---

### Common Applications

1. **Variance Comparison**
   - To check whether two samples have equal variances before applying a t-test.
   - Used in **preliminary analysis**.

2. **ANOVA (Analysis of Variance)**
   - F-test is the core test in ANOVA to determine if group means are significantly different.

3. **Regression Analysis**
   - Used to test the **overall significance** of a regression model.
   - H₀: All regression coefficients are zero.

---

### Example – Comparing Variances

Suppose we have two datasets:
- Sample A: variance = 25
- Sample B: variance = 10

**F = 25 / 10 = 2.5**

Using the F-distribution table, we determine if this ratio is statistically significant at a given α level (e.g., 0.05) with appropriate degrees of freedom.

---

### Conclusion

The **F-test** is a fundamental part of many statistical procedures, especially **ANOVA** and **regression**. It helps in assessing whether the observed differences in variance or means are **statistically significant**, guiding decisions in hypothesis testing.

```python
# Q: Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): Outcome, must be 0 or 1
    p (float): Probability of success (1), where 0 ≤ p ≤ 1

    Returns:
    float: Probability corresponding to outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1")

    return p if x == 1 else (1 - p)

# Example usage
print("P(X=1):", bernoulli_pdf(1, 0.6))  # Should return 0.6
print("P(X=0):", bernoulli_pdf(0, 0.6))  # Should return 0.4

# This function manually implements the PDF of a Bernoulli distribution.
# A Bernoulli distribution has two possible outcomes: 1 (success) with probability p, and 0 (failure) with probability 1-p.
# The function takes an outcome (x) and success probability (p), checks for valid input, and returns the corresponding probability.
```

```python
# Q: Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a given dataset.

    Parameters:
    data (list or numpy array): Input numerical dataset.

    Returns:
    numpy array: Array of Z-scores.
    """
    mean = np.mean(data)
    std_dev = np.std(data)

    if std_dev == 0:
        raise ValueError("Standard deviation is zero; Z-scores cannot be computed.")

    z_scores = (data - mean) / std_dev
    return z_scores

# Example dataset
data = np.random.normal(50, 10, 100)  # Simulated dataset with mean=50, std=10

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.figure(figsize=(8, 4))
plt.plot(z_scores, marker='o', linestyle='', color='blue')
plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
plt.xlabel("Index")
plt.ylabel("Z-score")
plt.title("Z-scores of Dataset")
plt.grid(True)
plt.legend()
plt.show()

# This function calculates Z-scores by subtracting the mean from each data point and dividing by the standard deviation.
# Z-scores indicate how many standard deviations a data point is from the mean.
# The plot shows Z-scores scattered across the index, with a horizontal line at Z=0 representing the mean.
```

```python
# Q: Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt

# STEP 1: Create a non-normal population
# We use the exponential distribution, which is skewed (not symmetric, not bell-shaped)
# This represents a population where small values are common, and large values are rare.
population = np.random.exponential(scale=2.0, size=100000)

# STEP 2: Apply the Central Limit Theorem
# We'll take many random samples from the population and calculate their average (mean).
# According to CLT, the distribution of these averages should look more normal (bell-shaped).

sample_size = 30       # Each sample will have 30 values
num_samples = 1000     # We'll take 1000 such samples
sample_means = []      # This will store the average of each sample

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))  # Calculate and store the sample mean

# STEP 3: Visualize the original population distribution
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='orange', edgecolor='black', density=True)
plt.title("Original Population (Exponential Distribution)")
plt.xlabel("Value")
plt.ylabel("Density")

# Explanation:
# This plot shows that our population is not normally distributed.
# It is skewed to the right, with a long tail of larger values.

# STEP 4: Visualize the distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='green', edgecolor='black', density=True)
plt.title("Distribution of Sample Means (CLT in Action)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")

# Explanation:
# Although the original population was not normal,
# the sample means form a bell-shaped (normal) distribution.
# This is the Central Limit Theorem: no matter the original shape, 
# the averages of many random samples tend to form a normal distribution.

plt.tight_layout()
plt.show()
```

```python
# Q: Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():
    """
    This function calculates and plots the standard normal distribution.
    Standard normal distribution is a special case of the normal distribution with:
    - Mean (μ) = 0
    - Standard deviation (σ) = 1
    It is used extensively in statistics for calculating probabilities and Z-scores.
    """

    # Define the range of x values (typically from -4 to 4 to capture almost all the data)
    x = np.linspace(-4, 4, 1000)

    # Calculate the PDF (Probability Density Function) for the standard normal distribution
    pdf = norm.pdf(x, loc=0, scale=1)  # loc=mean=0, scale=std=1

    # Plotting the standard normal distribution
    plt.figure(figsize=(10, 5))
    plt.plot(x, pdf, color='blue', linewidth=2, label='Standard Normal PDF')
    plt.fill_between(x, pdf, color='lightblue', alpha=0.5)

    # Adding mean and ±1σ, ±2σ, ±3σ lines for reference
    for z in [-3, -2, -1, 0, 1, 2, 3]:
        plt.axvline(z, color='gray', linestyle='--', linewidth=1)
        plt.text(z, norm.pdf(z) + 0.01, f'z={z}', ha='center', fontsize=9)

    # Customize the plot
    plt.title("Standard Normal Distribution (μ = 0, σ = 1)")
    plt.xlabel("Z-value")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Call the function to display the standard normal curve
plot_standard_normal_distribution()

# -----------------------------
# Explanation:
# -----------------------------
# - The function defines a range of Z-values from -4 to 4 (as ~99.99% of data falls in this range).
# - It uses scipy's norm.pdf to calculate the probability density at each point.
# - Vertical dashed lines are added at every integer Z-value for visual reference.
# - This plot visually represents how probability is distributed around the mean in a standard normal curve.
# - The highest point of the curve is at Z = 0, and it symmetrically decreases as we move away from the mean.
```

```python
# Q: Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

# Step 1: Define the known population parameters
population_mean = 100         # μ (known mean of the population)
population_std = 15           # σ (known standard deviation of the population)

# Step 2: Define the sample data
sample = [110, 112, 108, 115, 109, 111, 113, 107, 114, 116]
sample_size = len(sample)
sample_mean = np.mean(sample)

# Step 3: State the hypothesis
# H0 (Null Hypothesis): The sample comes from a population with mean = 100
# H1 (Alternative Hypothesis): The sample comes from a population with mean ≠ 100 (two-tailed test)

# Step 4: Calculate the Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 5: Set the significance level and find the critical Z-value
alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)  # two-tailed test

# Step 6: Decision rule
reject_null = abs(z_stat) > z_critical

# Step 7: Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"Critical Z-value: ±{z_critical:.4f}")
print("Result: ", "Reject Null Hypothesis (Significant Difference)" if reject_null else "Fail to Reject Null Hypothesis (No Significant Difference)")

# ---------------------------------------------------
# Explanation:
# ---------------------------------------------------
# - We are testing whether our sample mean significantly differs from the population mean using Z-statistics.
# - Z-statistic measures how far the sample mean is from the population mean in terms of standard errors.
# - A critical Z-value (for α = 0.05) is ±1.96 for a two-tailed test.
# - If the absolute Z-statistic is greater than 1.96, we reject the null hypothesis.
# - This method is appropriate when population standard deviation is known and sample size is sufficiently large (or population is normal).
```

```python
# Q: Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy import stats

# Step 1: Sample data
data = [52, 55, 60, 57, 58, 59, 61, 63, 62, 64]  # Sample dataset
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
n = len(data)

# Step 2: Set confidence level (e.g., 95%)
confidence = 0.95
alpha = 1 - confidence

# Step 3: Calculate the critical t-value
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Step 4: Calculate the margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Step 5: Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Step 6: Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Confidence Level: {confidence*100:.0f}%")
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# ---------------------------------------------------------
# Explanation:
# ---------------------------------------------------------
# - This is a 95% confidence interval for the true population mean.
# - It is calculated using the sample mean ± margin of error (based on t-distribution).
# - Since the population standard deviation is unknown and the sample size is small, we use the t-distribution.
# - The result means: "We are 95% confident that the true population mean lies between the lower and upper bounds."
# - This interpretation assumes random sampling and a roughly normal distribution of the sample data.
```

```python
# Q: Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_skewness_kurtosis(data):
    """
    Function to calculate and return skewness and kurtosis of a given dataset.
    
    Parameters:
    - data: A NumPy array or list of numerical values
    
    Returns:
    - skew_value: Measure of asymmetry in the distribution
    - kurtosis_value: Measure of tailedness of the distribution
    """
    skew_value = skew(data)
    kurtosis_value = kurtosis(data)  # By default, this returns excess kurtosis (kurtosis - 3)
    return skew_value, kurtosis_value

# Step 1: Generate synthetic normal data
np.random.seed(1)
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Call the function and print results
skewness, excess_kurtosis = analyze_skewness_kurtosis(normal_data)

print(f"Skewness: {skewness:.4f}")
print(f"Excess Kurtosis: {excess_kurtosis:.4f}")

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - **Skewness** measures the symmetry of the distribution:
#     - Skewness ≈ 0 → Symmetric (like normal distribution)
#     - Skewness > 0 → Right-skewed
#     - Skewness < 0 → Left-skewed
#
# - **Kurtosis** measures the "tailedness":
#     - Excess Kurtosis ≈ 0 → Normal tailedness
#     - > 0 → Heavy tails (leptokurtic)
#     - < 0 → Light tails (platykurtic)
#
# - A perfect normal distribution has skewness = 0 and excess kurtosis = 0.
# - These values are important for assessing normality assumptions in statistics.
```

```python
# Q: Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

# Step 1: Define the known population parameters
population_mean = 100      # Known population mean (μ)
population_std = 15        # Known population standard deviation (σ)

# Step 2: Generate a sample dataset
np.random.seed(10)
sample = np.random.normal(loc=102, scale=15, size=50)  # Sample with different mean
sample_mean = np.mean(sample)
sample_size = len(sample)

# Step 3: Perform the Z-test
# Formula: z = (x̄ - μ) / (σ / √n)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 4: Calculate the p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Step 5: Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 6: Interpretation
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis — the sample mean is significantly different from the population mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis — no significant difference between sample and population mean.")

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - A Z-test compares the sample mean to a known population mean when the population standard deviation is known.
# - The Z-score measures how many standard errors the sample mean is away from the population mean.
# - A low p-value (< 0.05) indicates strong evidence against the null hypothesis (i.e., the means are significantly different).
# - This test is useful when analyzing whether observed data deviates significantly from a known or expected value.
```

```python
# Q: Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.

import numpy as np
from scipy import stats

# Step 1: Generate two independent random samples
np.random.seed(42)
sample1 = np.random.normal(loc=100, scale=10, size=50)  # Mean = 100, Std = 10
sample2 = np.random.normal(loc=105, scale=10, size=50)  # Mean = 105, Std = 10

# Step 2: Perform a two-sample t-test (independent samples)
# Null Hypothesis (H0): There is no difference in means (mean1 = mean2)
# Alternative Hypothesis (H1): There is a significant difference (mean1 ≠ mean2)
t_statistic, p_value = stats.ttest_ind(sample1, sample2)

# Step 3: Display the results
print(f"Sample 1 Mean: {np.mean(sample1):.2f}")
print(f"Sample 2 Mean: {np.mean(sample2):.2f}")
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpretation
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis — the difference in means is statistically significant.")
else:
    print("Conclusion: Fail to reject the null hypothesis — no significant difference between the sample means.")

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - We simulated two random datasets from normal distributions with different means.
# - We used a two-sample t-test to check if their means are significantly different.
# - The t-test returns a t-statistic and a p-value.
# - The p-value helps us decide whether the observed difference is likely due to chance.
# - If the p-value is less than 0.05, we conclude that the difference in means is statistically significant.
```

```python
# Q: Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Define known population parameters
population_mean = 100     # μ
population_std = 15       # σ
alpha = 0.05              # Significance level (for two-tailed test)

# Step 2: Generate sample data
np.random.seed(0)
sample = np.random.normal(loc=105, scale=15, size=40)  # Sample with mean ≠ population mean
sample_mean = np.mean(sample)
sample_size = len(sample)

# Step 3: Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 4: Calculate critical Z values for two-tailed test
z_critical = norm.ppf(1 - alpha/2)  # two-tailed → split alpha in half
lower_critical = -z_critical
upper_critical = z_critical

# Step 5: Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Step 6: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Z-values: ±{z_critical:.2f}")

# Step 7: Visualize the decision region
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution', color='black')

# Fill rejection regions
plt.fill_between(x, y, where=(x <= lower_critical), color='red', alpha=0.5, label='Rejection Region (α/2)')
plt.fill_between(x, y, where=(x >= upper_critical), color='red', alpha=0.5)

# Plot observed Z-score
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.axvline(lower_critical, color='darkred', linestyle='--', label=f'Z = {lower_critical:.2f}')
plt.axvline(upper_critical, color='darkred', linestyle='--', label=f'Z = {upper_critical:.2f}')

# Formatting the plot
plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - We simulate sample data from a normal distribution with mean ≠ population mean.
# - A two-tailed Z-test is used to test if the sample mean is significantly different (in either direction).
# - Rejection regions are defined by ±Z_critical (based on α/2).
# - If the observed Z-score lies beyond these critical values, we reject the null hypothesis.
# - Visualization shows where the observed Z-score lies relative to the rejection zones.
```

```python
# Q: Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05):
    """
    Visualizes Type 1 and Type 2 errors for hypothesis testing between null and alternative hypotheses.
    
    Parameters:
    - mu0: Mean under the null hypothesis (H0)
    - mu1: Mean under the alternative hypothesis (H1)
    - sigma: Standard deviation (assumed known)
    - n: Sample size
    - alpha: Significance level (default 0.05)
    """

    # Step 1: Calculate standard error
    se = sigma / np.sqrt(n)

    # Step 2: Critical value for the significance level (two-tailed test)
    z_critical = norm.ppf(1 - alpha/2)

    # Step 3: Calculate critical sample mean boundaries under H0
    lower_bound = mu0 - z_critical * se
    upper_bound = mu0 + z_critical * se

    # Step 4: Define x range for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Step 5: Define PDFs under H0 and H1
    h0_pdf = norm.pdf(x, loc=mu0, scale=se)
    h1_pdf = norm.pdf(x, loc=mu1, scale=se)

    # Step 6: Plot distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, h0_pdf, label='H0: μ = {}'.format(mu0), color='blue')
    plt.plot(x, h1_pdf, label='H1: μ = {}'.format(mu1), color='green')

    # Step 7: Shade Type 1 Error region (α) — rejection region of H0
    plt.fill_between(x, h0_pdf, where=((x < lower_bound) | (x > upper_bound)),
                     color='red', alpha=0.3, label='Type I Error (α)')

    # Step 8: Shade Type 2 Error region (β) — fail to reject H0 when H1 is true
    plt.fill_between(x, h1_pdf, where=((x >= lower_bound) & (x <= upper_bound)),
                     color='orange', alpha=0.3, label='Type II Error (β)')

    # Annotations and formatting
    plt.axvline(lower_bound, color='black', linestyle='--', label='Critical values')
    plt.axvline(upper_bound, color='black', linestyle='--')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function
visualize_type1_type2()

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - Type I Error (α): Rejecting the null hypothesis when it is actually true. 
#   → This is shaded in **red** (outer tails of H0 distribution).
# - Type II Error (β): Failing to reject the null hypothesis when the alternative is true.
#   → This is shaded in **orange** (central region of H1 distribution).
# - Critical values divide the acceptance and rejection zones.
# - We use the standard normal distribution and sample size to scale the critical thresholds.
```

```python
# Q: Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind

# Step 1: Simulate two independent sample datasets
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Sample 1 (e.g., Treatment group)
group2 = np.random.normal(loc=55, scale=12, size=30)  # Sample 2 (e.g., Control group)

# Step 2: Perform an independent (two-sample) T-test
t_stat, p_value = ttest_ind(group1, group2, equal_var=False)  # Welch’s t-test

# Step 3: Print the results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis — the two group means are significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis — no significant difference between the two group means.")

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - An independent t-test compares the means of two independent groups.
# - The null hypothesis (H0): group1 mean = group2 mean.
# - The alternative hypothesis (H1): group1 mean ≠ group2 mean.
# - Welch's t-test is used when variances are not assumed equal.
# - If the p-value is less than 0.05, we conclude that there is a statistically significant difference.
```

```python
# Q: Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# Step 1: Simulate paired sample data (e.g., before and after treatment scores)
np.random.seed(0)
before = np.random.normal(loc=70, scale=8, size=30)    # Scores before intervention
after = before + np.random.normal(loc=3, scale=5, size=30)  # Scores after intervention

# Step 2: Perform paired sample T-test
t_stat, p_value = ttest_rel(before, after)

# Step 3: Print test results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis — significant difference between before and after.")
else:
    print("Conclusion: Fail to reject the null hypothesis — no significant difference detected.")

# Step 5: Visualization of before vs after
plt.figure(figsize=(10, 5))
x = np.arange(len(before))
plt.plot(x, before, label='Before', marker='o', linestyle='--', color='blue')
plt.plot(x, after, label='After', marker='s', linestyle='-', color='green')
plt.fill_between(x, before, after, color='gray', alpha=0.2)
plt.title('Paired Sample Comparison (Before vs After)')
plt.xlabel('Sample Index')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - A paired sample t-test compares the means of two related groups.
# - It is often used in pre-test/post-test scenarios where measurements are taken from the same subjects.
# - The null hypothesis assumes no difference between the two conditions.
# - A significant p-value (< 0.05) means that the intervention (or treatment) likely had an effect.
# - The line plot visually shows how each individual's score changed before and after.
```

```python
# Q: Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
from scipy.stats import ttest_1samp, norm

# Step 1: Simulate a sample dataset
np.random.seed(1)
sample = np.random.normal(loc=52, scale=10, size=30)  # Sample with mean=52, std=10, size=30
population_mean = 50
population_std = 10

# Step 2: Perform Z-test (when population std is known)
sample_mean = np.mean(sample)
sample_size = len(sample)
standard_error = population_std / np.sqrt(sample_size)
z_stat = (sample_mean - population_mean) / standard_error
p_value_z = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed

# Step 3: Perform one-sample T-test (when population std is unknown)
t_stat, p_value_t = ttest_1samp(sample, population_mean)

# Step 4: Print results
print("Z-test results:")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value_z:.4f}")
print()

print("T-test results:")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value_t:.4f}")
print()

# Step 5: Interpretation
alpha = 0.05
if p_value_z < alpha:
    print("Z-test: Reject the null hypothesis — sample mean is significantly different from population mean.")
else:
    print("Z-test: Fail to reject the null hypothesis — no significant difference.")

if p_value_t < alpha:
    print("T-test: Reject the null hypothesis — sample mean is significantly different from population mean.")
else:
    print("T-test: Fail to reject the null hypothesis — no significant difference.")

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - Z-test is used when the population standard deviation is known.
# - T-test is preferred when the population standard deviation is unknown and the sample size is small (< 30–40).
# - Both tests test the same null hypothesis but with different assumptions.
# - With larger sample sizes, T-test and Z-test results tend to converge.
# - This script helps understand the practical difference between both approaches.
```

```python
# Q: Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for the mean of a given sample.

    Parameters:
    - data: List or NumPy array of sample observations.
    - confidence: Confidence level (default is 0.95 for 95%).

    Returns:
    - A tuple with (lower_bound, upper_bound) of the confidence interval.
    """
    sample_mean = np.mean(data)                    # Step 1: Compute sample mean
    sample_std = np.std(data, ddof=1)              # Step 2: Sample standard deviation with Bessel's correction
    n = len(data)                                   # Step 3: Sample size
    standard_error = sample_std / np.sqrt(n)       # Step 4: Standard Error of the Mean (SEM)
    
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)  # Step 5: T-critical value for given confidence and df
    margin_of_error = t_critical * standard_error  # Step 6: Margin of error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return (lower_bound, upper_bound)

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=75, scale=10, size=30)
ci = confidence_interval(sample_data, confidence=0.95)

print(f"95% Confidence Interval for the sample mean: {ci}")

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - The **confidence interval** gives a range that is likely to contain the true population mean.
# - A 95% confidence level means: "We are 95% confident the interval contains the true mean."
# - This interval accounts for:
#   → Sample variability (standard deviation),
#   → Sample size (larger samples give narrower intervals),
#   → Desired confidence level (higher confidence gives wider intervals).
# - The t-distribution is used because population std deviation is unknown and sample size is relatively small.
# - This is useful for statistical reporting, uncertainty estimation, and inference.
```

```python
# Q: Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    """
    Calculates the margin of error for the mean of a sample.

    Parameters:
    - data: List or NumPy array of sample values.
    - confidence: Confidence level (default is 0.95 for 95%).

    Returns:
    - Margin of error value.
    """
    sample_std = np.std(data, ddof=1)          # Step 1: Sample standard deviation (with Bessel's correction)
    n = len(data)                               # Step 2: Sample size
    standard_error = sample_std / np.sqrt(n)   # Step 3: Standard error of the mean (SEM)

    t_critical = t.ppf((1 + confidence) / 2, df=n-1)  # Step 4: t-critical value from t-distribution
    margin = t_critical * standard_error        # Step 5: Margin of error calculation

    return margin

# Example usage
np.random.seed(10)
sample_data = np.random.normal(loc=100, scale=15, size=40)
moe = margin_of_error(sample_data, confidence=0.95)

print(f"Margin of Error (95% confidence): {moe:.4f}")

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - **Margin of Error (MOE)** quantifies the uncertainty around the sample mean.
# - It is the range above and below the sample mean where the true population mean is expected to lie.
# - A larger MOE means less precision; smaller MOE means more precise estimation.
# - The t-distribution is used because population standard deviation is unknown.
# - This is crucial in confidence intervals and survey reporting to express accuracy.
```

```python
# Q: Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

# In this example, we use Bayes' Theorem to update the probability of a person having a disease 
# given a positive test result.

def bayes_theorem(prior_A, prob_B_given_A, prob_B_given_not_A):
    """
    Applies Bayes' Theorem to compute the posterior probability of event A given event B.

    Parameters:
    - prior_A: P(A), the prior probability of event A (e.g., having the disease)
    - prob_B_given_A: P(B|A), the probability of B given A (e.g., test is positive if person has disease)
    - prob_B_given_not_A: P(B|¬A), the probability of B given not A (e.g., false positive rate)

    Returns:
    - posterior: P(A|B), the probability of A given B (e.g., person has disease given positive test)
    """
    # Step 1: Compute the probability of B
    prob_not_A = 1 - prior_A
    prob_B = (prob_B_given_A * prior_A) + (prob_B_given_not_A * prob_not_A)

    # Step 2: Apply Bayes' Theorem
    posterior = (prob_B_given_A * prior_A) / prob_B
    return posterior

# Example case:
# - 1% of people have a disease (prior)
# - Test is 95% accurate for detecting the disease (true positive rate)
# - Test has a 5% false positive rate

prior_disease = 0.01
true_positive_rate = 0.95
false_positive_rate = 0.05

posterior_probability = bayes_theorem(prior_disease, true_positive_rate, false_positive_rate)

print(f"Probability that a person has the disease given a positive test result: {posterior_probability:.4f}")

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - **Bayes' Theorem** helps us revise the probability of a hypothesis when given new evidence.
# - In real life, even if a test is accurate, the probability of actually having a rare disease after testing positive
#   may still be low — this is a classic example of conditional probability.
# - Bayes' formula is: P(A|B) = [P(B|A) * P(A)] / P(B)
#   → P(A|B): Probability of disease given positive test (posterior)
#   → P(B|A): Probability of positive test given disease (true positive)
#   → P(A): Prior probability of disease
#   → P(B): Total probability of positive test (true + false positives)
# - This method is widely used in medical testing, spam filtering, diagnostics, and machine learning (e.g., Naive Bayes).
```

```python
# Q: Perform a Chi-square test for independence between two categorical variables in Python.

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table (e.g., Gender vs. Preference)
# Suppose we have data on people's preference by gender
# Rows: Gender (Male, Female)
# Columns: Preference (Product A, Product B)

data = {'Product A': [30, 20], 'Product B': [10, 40]}
table = pd.DataFrame(data, index=['Male', 'Female'])

# Step 2: Apply Chi-square test for independence
chi2, p, dof, expected = chi2_contingency(table)

# Step 3: Output the results
print("Chi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p)
print("\nExpected Frequencies Table:")
print(pd.DataFrame(expected, index=table.index, columns=table.columns))

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - A **Chi-square test for independence** checks if two categorical variables are related.
# - H₀ (null hypothesis): The two variables are independent (no association).
# - H₁ (alternative hypothesis): The variables are dependent (associated).
# - The test compares observed values with expected values (if variables were truly independent).
# - If the p-value < 0.05 (significance level), we reject H₀ → indicating a statistically significant relationship.
# - Here, we used a 2x2 contingency table with counts of preferences by gender.
# - The `chi2_contingency` function handles the test and returns:
#   → chi2 statistic
#   → p-value
#   → degrees of freedom (dof)
#   → expected frequencies under independence assumption.
```

```python
# Q: Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Define the observed data as a contingency table
# Example: Survey data of preferences among two age groups
observed_data = {'Option 1': [25, 30], 'Option 2': [15, 20]}
observed_table = pd.DataFrame(observed_data, index=['Under 30', '30 and Above'])

# Step 2: Use chi2_contingency to compute expected frequencies
chi2, p, dof, expected_frequencies = chi2_contingency(observed_table)

# Step 3: Display expected frequencies
expected_table = pd.DataFrame(expected_frequencies, index=observed_table.index, columns=observed_table.columns)
print("Observed Frequencies:\n", observed_table)
print("\nExpected Frequencies:\n", expected_table)

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - **Expected frequencies** represent the counts we would expect in each cell 
#   of the contingency table if the two variables were truly independent.
# - The formula for each expected frequency is:
#     E = (row_total * column_total) / grand_total
# - `chi2_contingency()` from SciPy automatically calculates the expected table.
# - This is the foundational step before calculating the Chi-square statistic.
# - If observed values differ significantly from expected ones, it may suggest a relationship.
```

```python
# Q: Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
from scipy.stats import chisquare

# Step 1: Define observed frequencies
# Example: Observed dice rolls
observed = np.array([18, 22, 20, 17, 19, 24])  # Frequencies for each face from 1 to 6

# Step 2: Define expected frequencies (uniform distribution for a fair die)
expected = np.array([20] * 6)  # Expecting 20 occurrences of each face if the die is fair

# Step 3: Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Output the results
print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_value)

# -------------------------------------------------------------
# Explanation:
# -------------------------------------------------------------
# - **Goodness-of-fit test** checks if observed data matches an expected distribution.
# - H₀ (null hypothesis): Observed distribution = Expected distribution (e.g., fair die).
# - H₁ (alternative): Observed ≠ Expected.
# - If the p-value is less than 0.05, we reject the null hypothesis and conclude that the
#   observed frequencies differ significantly from the expected ones.
# - In this case, we're comparing dice roll outcomes to a uniform distribution using `scipy.stats.chisquare`.
# - The test returns the chi-square statistic and the associated p-value for inference.
```
